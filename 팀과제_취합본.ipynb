## 5) 하루 평균 인기동영상

```python
KRvideos["trending_date"] = pd.to_datetime(KRvideos["trending_date"], format = "%y.%d.%m")
```

```python
plt.figure(figsize = (15,10));
KRvideos["trending_date"].value_counts().plot()
plt.legend(prop = {"size":30});
```

```python
print("평균 : {:.2f}개".format(KRvideos["trending_date"].value_counts().mean()))
print("최대 : {:.2f}개".format(KRvideos["trending_date"].value_counts().max()))
print("최소 : {:.2f}개".format(KRvideos["trending_date"].value_counts().min()))
```

```python
JPvideos["trending_date"] = pd.to_datetime(JPvideos["trending_date"], format = "%y.%d.%m")
```

```python
plt.figure(figsize = (15,10));
JPvideos["trending_date"].value_counts().plot()
plt.legend(prop = {"size":30});
```

```python
print("평균 : {:.2f}개".format(JPvideos["trending_date"].value_counts().mean()))
print("최대 : {:.2f}개".format(JPvideos["trending_date"].value_counts().max()))
print("최소 : {:.2f}개".format(JPvideos["trending_date"].value_counts().min()))
```

최대값 역시 200개와 199개로 큰 차이가 나지 않았으나, 최소의 경우 102개와 40개로 큰 차이를 보였다.
